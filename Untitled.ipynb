```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
image = Image.open('./cat.jpeg')
image.thumbnail((150, 150), Image.ANTIALIAS)

plt.imshow(image)
plt.show()
image = np.array(image)

gray = np.dot(image[...,:3], [0.299, 0.587, 0.114])

plt.imshow(gray, cmap='gray')
print(np.shape(gray))
tiger_lattice = nx.grid_graph([150, 150])

tiger_lattice_A = nx.to_numpy_array(tiger_lattice)

tiger_lattice_D = np.zeros(np.shape(tiger_lattice_A))

for i in range(len(tiger_lattice_A)):
    tiger_lattice_D[i][i] = np.sum(tiger_lattice_A[i])

tiger_lattice_L = tiger_lattice_D - tiger_lattice_A

tiger_w, tiger_v = np.linalg.eigh(tiger_lattice_L)
tiger_v = tiger_v.T
E = tiger_v[0:30]
print(np.shape(E))
proj = np.dot(np.reshape(gray, [1, 22500]), E.T)
recon = np.dot(proj, E)
plt.imshow(np.reshape(recon, [150, 150]), cmap='gray')
```
